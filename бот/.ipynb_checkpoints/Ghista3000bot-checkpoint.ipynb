```python
print (2)
import telegram
import config
import time
import telebot
import random
TOKEN = config.token
bot = telegram.Bot(TOKEN)
from time import localtime

keywords1 = ["артерия", "артерия эластического типа", "эластическая артерия", "артерии"]
keywords2 = ["вена", "вена мышечного типа", "мышечная вена", "вены"]
keywords3 = ["сердце", "эндокард", "миокард", "сердца"]
keywords4 = ["костный мозг", "красный", "красного", "мозга"]
keywords5 = ["тимус", "вилочковая", "зобная", "тимуса"]
details1 = 'Артерия эластического типа\nОкраска орсеином\n1 - Эластические мембраны'
details2 = 'Вена мышечного типа\nОкраска гематоксилином и эозином\n1 - Эндотелий\n2 - Субэндотелий\n3 - Мышечный слой'
details3 = 'Эндокард, миокард и волокна Пуркинье\nОкраска гематоксилином и эозином\nІ – эндокард\nІІ – миокард\nІІІ – эпикард\n 1 - Волокна Пуркинье\n2 - Эндотелиальный слой эндокарда\n3 - Субэндотелиальный слой эндокарда\n4 - Мышечно-эластический слой эндокарда\n5 - Сединительнотканный слой эндокарда\n6 - Кардиомициты'
details4 = 'Красный костный мозг\nОкраска гематоксилином и эозином\nХ400'
details5 = 'Тимус (вилочковая, или зобная железа)\nОкраска гематоксилином и эозином, Х40\n* - Дольки тимуса\n1 - Соединительнотканная капсула\n2 - Соединительнотканные перегородки\n3 - Корковое вещество\n4 - Мозговое вещество\n5 - Тельца Гассаля'



def get_last_update_id(updates):
    id_list = list()
    for update in updates:
        id_list.append(update["update_id"])
    return(max(id_list))


def intro ():
    last_update_id = None
    learning = False
    while True:
        while learning == False:
            updates = bot.getUpdates(last_update_id, timeout=100)
            if len(updates) > 0:
                last_update_id = get_last_update_id(updates) + 1
                for update in updates:
                    last_message = update["message"]
                    last_message_text = last_message['text']
                    last_chat_id =  last_message['chat']['id']
                    if last_message_text == "ботаем":
                        bot.sendMessage(last_chat_id, "Тогда начинаем!")
                        bot.sendMessage(last_chat_id, "Я буду кидать гистологические препараты, а ты пиши, что на них изображено с маленькой буквы.")
                        bot.sendMessage(last_chat_id, "То, что подписано цифрами я сам тебе расскажу))")
                        bot.sendMessage(last_chat_id, "Чтобы я кинул следующий, пиши: го")
                        asking = False
                        score = 0
                        tasks = list(range(1, 6))
                        random.shuffle(tasks)
                        tasknum = 0
                        learning = True
                    if last_message_text == "я отдыхать":
                        bot.sendMessage(last_chat_id, "До скорого!")
                        learning = 0
                    if learning == False:
                        bot.sendMessage(last_chat_id, "Привет студент! Ну что, поботаем гисту?")
                        bot.sendMessage(last_chat_id, "Если хочешь начать, напиши: ботаем")
                        bot.sendMessage(last_chat_id, "Если же хочешь закончить, напиши: я отдыхать")
        while learning == True:
            updates = bot.getUpdates(last_update_id, timeout=100)
            if len(updates) > 0:
                last_update_id = get_last_update_id(updates) + 1
                for update in updates:
                    last_message = update["message"]
                    last_message_text = last_message['text']
                    last_chat_id =  last_message['chat']['id']
                    if tasknum >= 5:
                        bot.sendMessage(last_chat_id, "У меня больше нет препаратов!")
                        bot.sendMessage(last_chat_id, "Ты смог ответить правильно " + str(score) + " из 5")
                        bot.sendMessage(last_chat_id, "На этом и закончим!")
                        learning = False
                    if last_message_text == "я отдыхать":
                        bot.sendMessage(last_chat_id, "До скорого!")
                        bot.sendMessage(last_chat_id, "Но гиста сама себя не выучит)))")
                        learning = False
                    if tasknum < 5:
                        if last_message_text == "го":
                            chat_id = last_message['chat']['id']
                            some_image = open(str(tasks[tasknum]) + '.jpg', 'rb') # открыли файл на чтение как бинарный
                            bot.sendPhoto(chat_id, some_image) # сначала адресат, потом файл
                            some_image.close()
                            tasknum += 1
                            asking = True
            updates = bot.getUpdates(last_update_id, timeout=100)
            if len(updates) > 0:
                last_update_id = get_last_update_id(updates) + 1
                for update in updates:
                    last_message = update["message"]
                    last_message_text = last_message['text']
                    last_chat_id =  last_message['chat']['id']
                    if asking == True:
                        if tasks[tasknum - 1] == 1:
                            if any(keyword in last_message_text for keyword in keywords1):
                                bot.sendMessage(last_chat_id, "Нойс! Правильно))")
                                bot.sendMessage(last_chat_id, "А вот и пояснения:")
                                bot.sendMessage(last_chat_id, str(details1))
                                score += 1
                                asking = False
                            else:
                                bot.sendMessage(last_chat_id, "Неправильно, открою тебе секрет, это:")
                                bot.sendMessage(last_chat_id, str(details1))
                                asking = False
                    if asking == True:
                        if tasks[tasknum - 1] == 2:
                            if any(keyword in last_message_text for keyword in keywords2):
                                bot.sendMessage(last_chat_id, "Нойс! Правильно))")
                                bot.sendMessage(last_chat_id, "А вот и пояснения:")
                                bot.sendMessage(last_chat_id, str(details2))
                                score += 1
                                asking = False
                            else:
                                bot.sendMessage(last_chat_id, "Неправильно, открою тебе секрет, это:")
                                bot.sendMessage(last_chat_id, str(details2))
                                asking = False
                    if asking == True:
                        if tasks[tasknum - 1] == 3:
                            if any(keyword in last_message_text for keyword in keywords3):
                                bot.sendMessage(last_chat_id, "Нойс! Правильно))")
                                bot.sendMessage(last_chat_id, "А вот и пояснения:")
                                bot.sendMessage(last_chat_id, str(details3))
                                score += 1
                                asking = False
                            else:
                                bot.sendMessage(last_chat_id, "Неправильно, открою тебе секрет, это:")
                                bot.sendMessage(last_chat_id, str(details3))
                                asking = False
                    if asking == True:
                        if tasks[tasknum - 1] == 4:
                            if any(keyword in last_message_text for keyword in keywords4):
                                bot.sendMessage(last_chat_id, "Нойс! Правильно))")
                                bot.sendMessage(last_chat_id, "А вот и пояснения:")
                                bot.sendMessage(last_chat_id, str(details4))
                                score += 1
                                asking = False
                            else:
                                bot.sendMessage(last_chat_id, "Неправильно, открою тебе секрет, это:")
                                bot.sendMessage(last_chat_id, str(details4))
                                asking = False
                    if asking == True:
                        if tasks[tasknum - 1] == 5:
                            if any(keyword in last_message_text for keyword in keywords5):
                                bot.sendMessage(last_chat_id, "Нойс! Правильно))")
                                bot.sendMessage(last_chat_id, "А вот и пояснения:")
                                bot.sendMessage(last_chat_id, str(details5))
                                score += 1
                                asking = False
                            else:
                                bot.sendMessage(last_chat_id, "Неправильно, открою тебе секрет, это:")
                                bot.sendMessage(last_chat_id, str(details5))
                                asking = False
                        
            


intro()
```
